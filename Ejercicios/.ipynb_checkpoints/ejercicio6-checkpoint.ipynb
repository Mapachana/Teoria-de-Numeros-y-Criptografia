**i) Calcula los convergentes de $\sqrt{p}$.**

```python
def convergentes(n):
    p = 0
    k = 1
    i = 0
    parte_entera_raiz = math.floor(math.sqrt(n))
    q = p+parte_entera_raiz-(p+parte_entera_raiz%k)//k
    
    a_ant = 1
    b_ant = 0
    a = q
    b = 1
    
    a0 = a
    
    print("A[{}]= {}, B[{}]= {}".format(i, a, i, b))
    
    i += 1
    p_aux = p
    p = q*k-p
    
    k_aux = k
    k = (n-p*p)//k_aux
    
    q=(p+parte_entera_raiz-(p+parte_entera_raiz)%k)//k
    
    aux=a
    a=q*a+a_ant
    a_ant=aux

    aux_2=b
    b=q*b+b_ant
    b_ant=aux_2
    
    print("A[{}]= {}, B[{}]= {}".format(i, a, i, b))

    i=i+1
    p_aux=p
    p=q*k-p
    k_aux=k
    k=(n-p*p)//k_aux
    q=(p+parte_entera_raiz-(p+parte_entera_raiz)%k)//k


    aux=a
    a=q*a+a_ant
    a_ant=aux
    aux_2=b
    b=q*b+b_ant
    b_ant=aux_2
    
    print("A[{}]= {}, B[{}]= {}".format(i, a, i, b))

    while k != 1:
        i=i+1
        p_aux=p
        k_aux=k
        p=q*k-p
        k=(n-p_aux*p_aux)//k_aux+2*q*p_aux-(q*q)*k_aux
        q=(p+a0-(p+a0)%k)//k
        aux=a
        a=q*a+a_ant
        a_ant=aux

        aux_2=b
        b=q*b+b_ant
        b_ant=aux_2
        
        #if k == 1:
        #    return True
        
        print("A[{}]= {}, B[{}]= {}".format(i, a, i, b))

```

**iii)Calcula las unidades del anillo de enteros cuadráticos $\mathbb{Z}[\sqrt{p}]$.**

Por tanto cualquier unidad del anillo cuadrático $\mathbb{Z}[\sqrt{p}]$ es una potencia salvo el signo de $a+b\sqrt{p}$, es decir:

$$x+y\sqrt{p}=±(a+b\sqrt{p})^n$$

**iv)  ¿ Es $\mathbb{Z}[\sqrt{p}]$ el anillo de enteros del cuerpo $\mathbb{Q}[\sqrt{p}]$?**

```python
p%4
```

Como $p\equiv 1 \mod 4$ tenemos que $O_{\mathbb{Q}[\sqrt{p}]}=\{ m+n\frac{1+\sqrt{p}}{2}, m,n\in\mathbb{Z}\} =\mathbb{Z}[\frac{1+\sqrt{p}}{2}]$, por lo que $\mathbb{Z}[\sqrt{p}]$ no es anillo de enteros de $\mathbb{Q}[\sqrt{p}]$.
